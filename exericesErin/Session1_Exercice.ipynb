Exercise 7 - Write a program to arrange the characters of a string so that all lowercase letters should come first.
<br>
Expected result = eonoieXXMBL

````python
s1 = 'eXXonMoBiLe'

lower = []
upper = []
for char in s1:
    if char.islower():
        # add lowercase characters to lower list
        lower.append(char)
    else:
        # add uppercase characters to lower list
        upper.append(char)

# Join both list
sorted_str = ''.join(lower + upper)
print('Result:', sorted_str)
````

FOMC Minutes Sentiment Analysis
<br>
FOMC minutes are the official written record of the discussions that take place during meetings of the
Federal Open Market Committee (FOMC), the monetary policy–setting body of the U.S. Federal Reserve.
<br>
<br>
The minutes summarize:
- Economic conditions discussed (inflation, growth, labor market)
- Risks to the economic outlook
- Views on interest rates and monetary policy
- Areas of agreement and disagreement among members

<br>

**Why are FOMC minutes important for markets?**

<br>

Investors analyze the minutes to:
- Infer the future path of interest rates
- Assess whether the Fed is:
    - Hawkish (concerned about inflation, likely to raise rates)
    - Dovish (concerned about growth, likely to cut rates)

<br>

Even small wording changes can move:
- Bond yields
- Equity markets
- Exchange rates (USD)
