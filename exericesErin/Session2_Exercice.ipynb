Exercise 9 - Create the below functions
- Count the number of upper and lower case
- Count the number of occurrence of each character

````python
str1 = 'GeneralMotors'
str2 = 'ExxonMobile'

countL = 0
countU = 0

for char in str1:
    if char.islower():
        countL += 1
    elif char.isupper():
        countU += 1

for char in str2:
    if char.islower():
        countL += 1
    elif char.isupper():
        countU += 1

print("Lowercase letters:", countL)
print("Uppercase letters:", countU)

counts = {}

for char in str1 + str2:
    if char in counts:
        counts[char] += 1
    else:
        counts[char] = 1

print(counts)

````

Exercise 12 - Create a recursive function to compute a sum of all the number from 0 to n

````python
n = 10
def recursive_sum(n):
    if n == 0:
        return 0
    return n + recursive_sum(n - 1)
````
